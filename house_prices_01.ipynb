```python
import pandas as pd

from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
file_path = 'D:\Development\Python\melbone\melb_data.csv'
data = pd.read_csv(file_path)

data.describe()
```

```python
data.head()
```

```python
final_data = data.dropna(axis=0)
final_data.head()
```

```python
final_data.columns
```

```python
y = final_data.Price
model_features = ['Bathroom', 'Landsize','BuildingArea','YearBuilt','Lattitude','Longtitude']
X = final_data[model_features]

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)
```

Get Mean Absolute Error Method

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    pred = model.predict(val_X)
    mae = mean_absolute_error(val_y, pred)
    return(mae)
```

 Compare Accuracy of models built with different values for max_leaf_nodes

```python
my_dict = {}
candidate_max_leaf_nodes =  [5, 50, 500, 5000]

for max_leaf_nodes in candidate_max_leaf_nodes:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    my_dict[max_leaf_nodes] = my_mae
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))

best_tree_size = min(my_dict, key=my_dict.get)
print("The Best Tree Size is: ", best_tree_size)
```

Using the Minimum MAE Create Prediction Model

```python
data_model = DecisionTreeRegressor(max_leaf_nodes=best_tree_size, random_state=0)
data_model.fit(train_X, train_y)
final_prediction = data_model.predict(val_X)
print("The final prediction is: ", final_prediction)
```
